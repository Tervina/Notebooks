```r
myDataSet<-read.csv('G1_Allometry.csv')
myDataSet
myDataSet<-read.csv('G1_Allometry.csv',na.strings = c(''))
complete.cases(myDataSet)
myDataSet<-read.csv('G1_Allometry.csv',na.strings = c('NA'))
complete.cases(myDataSet)
myDataSet[!complete.cases(myDataSet),]
str(myDataSet)
summary(myDataSet)
```

```r

is.na(myDataSet)
any(is.na(myDataSet$height))
myDataSet$height=gsub(',','.',myDataSet$height)
myDataSet$height<-as.numeric(myDataSet$height)
summary(myDataSet)
```

```r
install.packages("mice")
library(mice)
colimputation<-mice(
myDataSet,
m=5,
method = c("","","pmm","",""),
maxit = 20
)
colimputation$imp
myNewDataSet<-complete(colimputation,5)
summary(myNewDataSet)
```

```r
install.packages("tidyr")
install.packages("tidyr")
library(tidyr)
cleanDataSet<-drop_na(myDataSet)
cleanDataSet
```

```r
install.packages("tidyverse")
library(tidyverse)
draw<-ggplot(cleanDataSet)
draw<-ggplot(cleanDataSet,aes(x=height,y=diameter))
draw+geom_point()
```

```r
draw_histo<-ggplot(cleanDataSet,aes(height))
draw_histo+geom_histogram()
draw_histo+geom_histogram(binwidth = 5)
draw_histo+geom_histogram(color="red")
```

```r
draw_bar<-ggplot(myDataSet,aes(x=diameter,fill=species))
draw_bar+geom_bar()
```

```r
draw_scater<-ggplot(myDataSet,aes(height,branchmass))
draw_scater
draw_scater+geom_point(aes(color=species))
draw_scater<-ggplot(cleanDataSet,aes(height,branchmass))
draw_scater+geom_point(aes(color=species))
draw_scater+geom_point(aes(color=species))+stat_smooth(se=TRUE)
```
